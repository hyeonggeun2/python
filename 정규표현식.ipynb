```python
html = ''
with open('webtoon_list.html', 'rt') as f:
    html = f.read()
```

```python
len(html)
```

```python
import re
re.findall(r'<a.*?class="title".*?>([\w\s]+)</a>', html) # 제목
```

```python
re.findall(r'<a\shref="(.*?)".*?class="title".*?>([\w\s]+)</a>', html) # 주소 및 제목
```

```python
re.findall(r'<a\s*href=".*?(\d+).*?".*?class="title".*?>([\w\s]+)</a>', html) # 고유번호 및 제목
```

```python
m_list = re.finditer(r'<a\s*href=".*?(?P<number>\d+).*?".*?class="title".*?>(?P<title>[\w\s]+)</a>', html) # 리스트에 값을 저장
```

```python
data = [m.group("title") for m in m_list] # 리스트중에서 이름만 뽑아서 data에 저장
```

```python
data
```

```python
d_count = {} # 중복개수 세는 알고리즘
for title in data:
    try: d_count[title]+= 1
    except: d_count[title]=1
print(d_count)
```

```python
import operator
collections= sorted(d_count.items(), key=lambda x:x[1])
```

```python
list(set(data)) # 중복을 제거한 리스트 (set은 중복을 가질 수 없으므로 중복이 제거됨)
```
