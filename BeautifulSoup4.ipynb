```python
def webtoon_info(title):
    result = [];
    soup = BeautifulSoup(open('webtoon_list.html', 'rt').read())
    toon_list = soup.select(f'a.title[title*="{title}"]')
    
    for toon in toon_list:
        title = toon.get_text()
        href = toon['href']
        thumbnail = toon.parent.select_one('img')['src']
    
        data = {
            'title': title,
            'href': href,
            'thumbnail': thumbnail
        }
    
        result.append(data)
    
    return result
```

```python
webtoon_info_dict = {}

soup = BeautifulSoup(open('webtoon_list.html', 'rt').read())
for toon in soup.select('a.title'):
    href = toon['href']
    title = toon['title']
    src = toon.parent.select_one('img')['src']
    
    webtoon_info_dict[title] = {
        'href': href,
        'src': src
    }
    
def webtoon_info(title):
    result_dict = {}
    for key, value in webtoon_info_dict.items():
        if title in key:
            result_dict[key] = value
    return result_dict
```

```python
res = {}
def show_weekday_webtoons():
    week = soup.select('div.daily_all')
    for day_list in week:
        day = day_list.select('ul')
        
    for d in day:
        day_name = d.parent.parent('h4')
        res_d = d.select('a.title')
        
        res_title = []
        for d_name in day_name:
            res_day = d_name.get_text()
        for r_d in res_d:
            res_title.append(r_d.get_text())
        
        res[res_day[0]] = res_title

show_weekday_webtoons()
```
