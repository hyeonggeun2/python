```python
import csv
import xml.etree.ElementTree as ET
from openpyxl import load_workbook
```

```python
# xml file
## xml food
doc = ET.parse('./food.xml')
root = doc.getroot()
datas = root.findall('Row')[1:]
## xml seafood
sea_doc = ET.parse('./seafood.xml')
sea_root = sea_doc.getroot()
sea_datas = sea_root.findall('Row')[1:]

# excel file
load_wb = load_workbook('./food.xlsx', data_only=True)
load_ws = load_wb['피벗그리드']

food_list = []

# custom list
add_list = ['돼지갈비', '목살', '삼겹살', '앞다리살', '갈비', '등심',
            '불고기', '한우안심', '한우양지', '대파', '쪽파', '꽈리고추',
           '청양고추', '애호박']

```

```python
# xml
for data in datas:
    food_list.append(data.find("HSK류코드").text)
## xml seafood
for data in sea_datas:
    food_list.append(data.find("수산물수출입품목명").text)

# execl
for name in load_ws['B']:
    if name.value != None:
        food_list.append(name.value)
```

```python
for item in add_list:
    food_list.append(item)
    
food_list = list(set(food_list))
```
